```python
# This code is to plot yearli average TOC values (from OMI and the NN method)
import matplotlib.pyplot as plt
import numpy as np

years   = np.array([2014, 2015, 2017, 2018, 2019])
NN_TOC  = np.array([322.70, 318.18, 305.80, 320.86, 323])
OMI_TOC = np.array([322.70, 320.15, 302.67, 319.80, 319.45])
LUT_TOC = np.array([323.42, 317.91, 303.26, 319.13, 321.68])

plt.scatter(years, NN_TOC)
plt.scatter(years, OMI_TOC)
plt.show()


```

```python
#rel. error function.
def relerror(x,y):
    z = x-y
    for i in x:
        error=z*100/i
    return error

nn_omi_error = relerror(OMI_TOC,NN_TOC)
lut_omi_error = relerror(OMI_TOC,LUT_TOC)

print(nn_omi_error)
print(lut_omi_error)


```
